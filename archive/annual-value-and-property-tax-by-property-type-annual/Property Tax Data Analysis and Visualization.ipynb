```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load data
df = pd.read_csv('annual-value-and-property-tax-by-property-type.csv')
df.replace('na', np.nan, inplace=True)
df['annual_value'] = pd.to_numeric(df['annual_value'])
df['median_annual_value'] = pd.to_numeric(df['median_annual_value'])
df['no_of_cases'] = pd.to_numeric(df['no_of_cases'])
df['property_tax_collection'] = pd.to_numeric(df['property_tax_collection'])

# 1. Price Trend / Tax Trend
# Let's calculate tax per case as an indicator of average property tax price
df['tax_per_case'] = df['property_tax_collection'] / df['no_of_cases']

# Save cleaned data to CSV for PowerBI
df.to_csv('cleaned_property_tax_data.csv', index=False)

print("Cleaned data saved to 'cleaned_property_tax_data.csv'")
print(df.head())
```

```python
# Correlation Matrix
corr_matrix = df[['no_of_cases', 'annual_value', 'median_annual_value', 'property_tax_collection', 'tax_per_case']].corr()
print("--- CORRELATION MATRIX ---")
print(corr_matrix)

# Plot Heatmap
plt.rcParams['figure.figsize'] = [8, 6]
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Correlation Heatmap of Property Tax Metrics')
plt.tight_layout()
plt.savefig('correlation_heatmap.png')
plt.close()
print("Saved correlation_heatmap.png")
```

```python
# Sort data by property type and year to ensure correct shift
df = df.sort_values(by=['property_type', 'financial_year']).reset_index(drop=True)

# Calculate YoY percentage changes
df['tax_yoy_return'] = df.groupby('property_type')['property_tax_collection'].pct_change() * 100
df['cases_yoy_return'] = df.groupby('property_type')['no_of_cases'].pct_change() * 100
df['median_av_yoy_return'] = df.groupby('property_type')['median_annual_value'].pct_change() * 100

print("--- SAMPLE RETURNS ---")
print(df[['financial_year', 'property_type', 'tax_yoy_return', 'cases_yoy_return']].dropna().head(10))

# Save the updated data with returns
df.to_csv('cleaned_property_tax_data_with_returns.csv', index=False)
print("Saved updated dataset to 'cleaned_property_tax_data_with_returns.csv'")
```

```python
# Plot Return Distribution
plt.rcParams['figure.figsize'] = [10, 6]
sns.histplot(data=df.dropna(subset=['tax_yoy_return']), x='tax_yoy_return', kde=True, bins=20)
plt.title('Distribution of Year-over-Year Property Tax Collection Returns (%)')
plt.xlabel('YoY Return (%)')
plt.ylabel('Frequency')
plt.tight_layout()
plt.savefig('tax_return_distribution.png')
plt.close()

# Distribution by Property Type using boxplot or multiple histograms
sns.boxplot(data=df.dropna(subset=['tax_yoy_return']), x='property_type', y='tax_yoy_return')
plt.title('YoY Property Tax Collection Returns by Property Type')
plt.xlabel('Property Type')
plt.ylabel('YoY Return (%)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('tax_return_distribution_by_type.png')
plt.close()

print("Saved distribution plots.")
```

```python
# Price Trend: Let's plot property_tax_collection over time by property type
plt.rcParams['figure.figsize'] = [10, 6]
sns.lineplot(data=df, x='financial_year', y='property_tax_collection', hue='property_type', marker='o')
plt.title('Property Tax Collection Trend (2002-2019)')
plt.xlabel('Financial Year')
plt.ylabel('Property Tax Collection (in thousands/units)')
plt.legend(title='Property Type')
plt.tight_layout()
plt.savefig('property_tax_collection_trend.png')
plt.close()

# Median Annual Value Trend (2010-2019)
sns.lineplot(data=df.dropna(subset=['median_annual_value']), x='financial_year', y='median_annual_value', hue='property_type', marker='s')
plt.title('Median Annual Value Trend (2010-2019)')
plt.xlabel('Financial Year')
plt.ylabel('Median Annual Value')
plt.legend(title='Property Type')
plt.tight_layout()
plt.savefig('median_annual_value_trend.png')
plt.close()

# Price Volume: Let's plot no_of_cases over time by property type
sns.lineplot(data=df, x='financial_year', y='no_of_cases', hue='property_type', marker='^')
plt.title('Volume Trend (Number of Cases, 2002-2019)')
plt.xlabel('Financial Year')
plt.ylabel('Number of Cases')
plt.legend(title='Property Type')
plt.tight_layout()
plt.savefig('volume_cases_trend.png')
plt.close()

print("Saved trend plots.")
```

```python
# Summary statistics of returns by property type
summary_stats = df.groupby('property_type')[['tax_yoy_return', 'cases_yoy_return', 'median_av_yoy_return']].mean()
print("--- AVERAGE YOY GROWTH RATES (%) BY PROPERTY TYPE ---")
print(summary_stats)

print("\n--- DETAILED CORRELATION FOR POWERBI HIGHLIGHTS ---")
for pt in df['property_type'].unique():
    pt_df = df[df['property_type'] == pt]
    corr = pt_df['no_of_cases'].corr(pt_df['property_tax_collection'])
    print(f"Correlation between Cases and Tax Collection for {pt}: {corr:.2f}")
```
