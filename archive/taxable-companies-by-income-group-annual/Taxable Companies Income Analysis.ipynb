```python
import pandas as pd

# Load the datasets
df1 = pd.read_csv('taxable-companies-by-assessed-income-group-ya2004-ya2007.csv')
df2 = pd.read_csv('taxable-companies-by-chargeable-income-group-ya2008-ya2019.csv')

# Inspect columns
print("Dataset 1 columns:", df1.columns)
print("Dataset 1 head:\n", df1.head())
print("\nDataset 2 columns:", df2.columns)
print("Dataset 2 head:\n", df2.head())
```

```python
# Rename columns to standardize
df1.columns = ['year', 'income_group', 'no_of_companies', 'total_income', 'net_tax']
df2.columns = ['year', 'income_group', 'no_of_companies', 'total_income', 'net_tax']

# Combine
df = pd.concat([df1, df2], ignore_index=True)

# 1. Price Trend (Total Income Trend)
# Group by year
trend_income = df.groupby('year')['total_income'].sum()

# 2. Price Volume (Number of Companies Trend)
trend_companies = df.groupby('year')['no_of_companies'].sum()

# 3. Correlation Heatmap
correlation_matrix = df[['no_of_companies', 'total_income', 'net_tax']].corr()

# 4. Return Distribution (Income Group Distribution)
# This is a bit ambiguous, I will show the distribution of companies per income group aggregated over all years
distribution = df.groupby('income_group')['no_of_companies'].sum().sort_values(ascending=False)

print("Income Trend:\n", trend_income)
print("\nCompany Trend:\n", trend_companies)
print("\nCorrelation Matrix:\n", correlation_matrix)
print("\nDistribution of Companies per Income Group:\n", distribution)
```
